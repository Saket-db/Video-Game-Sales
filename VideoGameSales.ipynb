```python
import pandas as pd

file_path = '/content/drive/MyDrive/Video Game Sales/VideoGamesSales.csv' #CSV in google drive
df = pd.read_csv(file_path)

print(df.shape) # Shape tells you about number of rows and columns does the data set
df.head()
```

```python
import numpy as np
df[df.duplicated()] # To know how many rows are duplicate
# df.shape[1]
```

```python
# current counts
print("rows now:", df.shape[0])
print("duplicates now:", df.duplicated().sum())
print("unique rows now (drop_duplicates):", df.drop_duplicates().shape[0])
```

```python
#Removing duplicate rows

df = df.drop_duplicates()
df.shape[0]
```

```python
#To check null Values
df = df.info()

#After running the command, found out that null values exist in 2 columns ie:Region,Publisher
```

```python
# Replace null Region values with "South"
df['Region'] = df['Region'].fillna('South')
df['Region'] = df['Region'].replace('Other', 'East')
df['Region'] = df['Region'].replace('Central', 'North')

# Verify no nulls remain
print("Nulls in 'Region':", df['Region'].isnull().sum())
```

```python
#Removing dollar sign from sales using regex
df['National Sales'] = (
    df['National Sales']
    .astype(str)               # ensure it's all strings
    .replace('[\$,]', '', regex=True)  # remove $ and commas
    .astype(float)             # convert to float
)
```

```python
# Fixing the problem of null values in publisher
df['Publisher'] = df['Publisher'].fillna('Unknown')
print('Nulls in Publishers:', df['Publisher'].isnull().sum())
# df.info()
```

All the null values have been filled and the dollar sign has been taken care off

```python
df['Country'] = df['Country'].replace({'USA':'United States'})
df['Country'] = df['Country'].str.title()

df=df.rename(columns={'NA_Sales':'National Sales','Global_Sales':'Global Sales','NA_Profit':'National Profit','Global_Profit':'Global Profit'})
```

```python
print(df['National Sales'].dtype)
sales_cap = df['National Sales'].quantile(0.95)
print("95th percentile value:", sales_cap)
print("min, max:", df['National Sales'].min(), df['National Sales'].max())

df['National Sales'] = np.where(df['National Sales'] > sales_cap, sales_cap, df['National Sales'])
df
```

```python
#Visualization using MatPlot and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

National_Sales = df.groupby(['Region', 'Country'])['National Sales'].sum().reset_index().sort_values(by='National Sales')
National_Sales
```

```python
#create a bar Chart

National_Sales=df.groupby(['Region','Country'])['National Sales'].sum().reset_index().sort_values(by='National Sales', ascending=False)
National_Sales
plt.figure(figsize=(16, 8))
sns.barplot(data=National_Sales, x='Region', y='National Sales',hue='Country')
plt.title('National Sales by Region and Country',fontweight='bold')
plt.xlabel('Region',fontweight='bold')
plt.ylabel('National Sales',fontweight='bold')
plt.show()
```

```python
#Create Box Plot
# ,meanprops={"marker":"o"
#                                                                                   ,"markerfacecolor":"white"
#                                                                                   ,"markersize":"3"
#                                                                                   ,"markeredgecolor":"green"}
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='National Sales', data=df,showmeans=True,hue='Genre',meanprops={"marker":"o"
                                                                                   ,"markerfacecolor":"white"
                                                                                   ,"markersize":"6"
                                                                                 ,"markeredgecolor":"green"})

plt.title('National Sales by Country',fontweight='bold')
plt.xlabel('Country',fontweight='bold')
plt.ylabel('National Sales',fontweight='bold')
plt.show()
```

```python
# Create a line chart
plt.figure(figsize=(10, 6))

# Line for National Sales
plt.plot(df['Year'], df['National Sales'], marker='o', linestyle='-', color='b', label='National Sales')

# Line for Global Sales
plt.plot(df['Year'], df['Global Sales'], marker='s', linestyle='--', color='g', label='Global Sales')

# Adding titles and labels
plt.title('National and Global Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()

# Display the line chart
plt.grid(True)
plt.show()
```
